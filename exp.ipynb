```python
def parse_entity_string(entity_string):
    if pd.isna(entity_string) or entity_string == '':
        return {}
    
    entities = {}
    pairs = entity_string.split(';')
    
    for pair in pairs:
        if ',' in pair:
            name = pair.split(',')[0].strip()  # remove numbers
            
            if name:
                if name in entities:
                    entities[name] += 1
                else:
                    entities[name] = 1
    
    return entities
```

```python
data["names_dict"] = data["persons"].apply(lambda X :parse_entity_string(X) )
data["organizations_dict"] =data["organizations"].apply(lambda X :parse_entity_string(X) )
```

```python
import pandas as pd

def inspect_model_features(dataset: NERDatasetV2, 
                           original_df: pd.DataFrame, 
                           tokenizer: AutoTokenizer, 
                           id2tag: dict, 
                           sample_index: int = 0):
    """
    Decodes and displays a single sample from the NER dataset for visual inspection.

    Args:
        dataset: The created NERDatasetV2 object.
        original_df: The DataFrame (train_df or test_df) that the dataset was created from.
        tokenizer: The tokenizer used for encoding.
        id2tag: The dictionary mapping label IDs back to tag strings.
        sample_index: The index of the sample to inspect.
    """
    
    print("="*80)
    print(f"INSPECTING SAMPLE AT INDEX: {sample_index}")
    print("="*80)

    # 1. Get the raw text and ground truth labels from the original DataFrame
    raw_data = original_df.iloc[sample_index]
    print(f"Original Text:\n{raw_data['text']}\n")
    print(f"Ground Truth Persons: {raw_data['persons']}")
    print(f"Ground Truth Orgs: {raw_data['organizations']}")
    print(f"Ground Truth Locs: {raw_data['locations']}\n")

    # 2. Get the processed features and labels from the dataset
    item = dataset[sample_index]
    input_ids = item['input_ids'].tolist()
    labels = item['labels'].tolist()
    
    # 3. Decode the tokens and labels
    decoded_tokens = tokenizer.convert_ids_to_tokens(input_ids)
    decoded_labels = []
    for label_id in labels:
        if label_id == -100:
            # -100 is the special value for tokens to be ignored by the loss function
            # (e.g., [CLS], [SEP], padding)
            decoded_labels.append("[IGNORE]")
        else:
            decoded_labels.append(id2tag[label_id])
            
    # 4. Combine into a DataFrame for pretty printing
    inspection_df = pd.DataFrame({
        'Token': decoded_tokens,
        'Label': decoded_labels,
        'Input ID': input_ids,
        'Label ID': labels
    })
    
    # Filter out padding tokens for cleaner display
    inspection_df = inspection_df[inspection_df['Token'] != tokenizer.pad_token]
    
    # Display the full DataFrame
    with pd.option_context('display.max_rows', None, 'display.max_columns', None, 'display.width', 1000):
        print("--- Model Input and Labels ---")
        print(inspection_df)
    
    # For a quick check, let's highlight only the tokens that were tagged as entities
    print("\n--- Highlight: Tokens Tagged as Entities ---")
    entity_tokens = inspection_df[~inspection_df['Label'].isin(['O', '[IGNORE]'])]
    if entity_tokens.empty:
        print("No entities were tagged for this sample.")
    else:
        with pd.option_context('display.max_rows', None, 'display.max_columns', None):
            print(entity_tokens)


# --- HOW TO USE IT ---

# Let's inspect the first 3 samples of our training data
for i in range(3):
    inspect_model_features(
        dataset=train_dataset,
        original_df=train_df,
        tokenizer=tokenizer,
        id2tag=id2tag,
        sample_index=i
    )
```
